```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# population de dpart - on ne peut pas calculer cette moyenne en gnrale
population = np.random.randint(1, 7, size=100_000)

"""
Calcul des moyennes de nos chantillons dans la population pop 
"""
def sample_means(pop, size, n):
    # chantillonnage avec remise les chantillons sont pris de manire alatoire et indpendant 
    return [ np.mean(np.random.choice(pop, size=size, replace=True )) for _ in range(n) ]

# Calculer les moyennes pour deux tailles d'chantillons

t_1 = 5
t_2 = 30

means_small = sample_means(pop = population, size = t_1, n = 1000)
means_large = sample_means(pop = population, size = t_2, n = 1000)


# Bilan  

### On ne peut en gnrale pas la calcul mais on le fait ici ... 
print("MOYENNE THEORIQUE ????", np.mean(population) ) 
###

print(f"MOYENNE PETIT ECHANTILLON N = 5" , np.mean(means_small ) )
print(f"MOYENNE GRAND ECHANTILLON N = 30" , np.mean(means_large ) )


# Visualisation
plt.figure(figsize=(14, 6))

# Histogramme des petits chantillons
plt.subplot(1, 2, 1)
sns.histplot(means_small, bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title(f'Taille des chantillons = {t_1}')
plt.xlabel('Moyenne des chantillons')
plt.ylabel('Frquence')

# Histogramme des grands chantillons
plt.subplot(1, 2, 2)
sns.histplot(means_large, bins=30, kde=True, color='orange', edgecolor='black')
plt.title(f'Taille des chantillons = {t_2}' )
plt.xlabel('Moyenne des chantillons')
plt.ylabel('Frquence')

plt.tight_layout()
plt.show()
```
